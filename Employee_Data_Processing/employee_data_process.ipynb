```python
gpt3["Bonus_norm"] = gpt3.groupby("Education")["Bonus"].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
gpt3.groupby("Department")["Bonus_norm"].mean().sort_values(ascending=False).head(2)
```

```python
gpt3["Agegrup"]=pd.cut(
    gpt3["Age"],
    bins=[0,29,40,50,float("inf")],
    labels=["Genc","Orta","Ortayasli","Yasli"]
)
gpt3.groupby("Agegrup",as_index=False).agg({
    "Salary":"mean",
    "PerformanceScore":"mean"
})
```

```python
import numpy as np
import pandas as pd
```

```python
gpt3["Isegirisyili"]=pd.to_datetime(
    np.random.choice(pd.date_range("01-01-2015","01-01-2025"),size=(len(gpt3))),format="%d-%m-%Y"
)
gpt3["Year"]=gpt3["Isegirisyili"].dt.year
pd.pivot_table(gpt3,index="Department",columns="Year",aggfunc="size")
```

```python
pd.pivot_table(gpt3,index="Department",columns="Year",values=["Salary","PerformanceScore"],aggfunc="mean")
```

```python
gpt3 = gpt3.sort_values(["Department", "Isegirisyili"])
gpt3["RollingSalary"] = gpt3.groupby("Department")["Salary"] \
                             .rolling(window=10, min_periods=1) \
                             .mean() \
                             .reset_index(level=0, drop=True)
pd.pivot_table(
    gpt3,
    index="Department",
    columns=gpt3["Isegirisyili"].dt.to_period("Y"),  # ay bazlı
    values="RollingSalary",
    aggfunc="first"  # zaten her hücrede 1 değer var
)
```

```python
gpt3["Performance_norm"]=(gpt3["PerformanceScore"]-gpt3["PerformanceScore"].min()) / (gpt3["PerformanceScore"].min()-gpt3["PerformanceScore"].max())
gpt3["Projects_norm"]=(gpt3["ProjectsCompleted"]-gpt3["ProjectsCompleted"].min()) / (gpt3["ProjectsCompleted"].min()-gpt3["ProjectsCompleted"].max())
gpt3["Performans_skor"]=gpt3["Performance_norm"]*0.6 + gpt3["Projects_norm"]*0.4
gpt3.groupby("Department",as_index=False)["Performans_skor"].mean()
```

```python
pt= pd.pivot_table(
    gpt3,
    index=["Department","Education"],
    columns="Remote",
    values=["Salary","Bonus"],
    aggfunc="mean",
    sort="Bonus"
)
pt.sort_values([("Bonus",True),("Bonus",False)],ascending=False)
```

```python
min_salary=gpt3["Salary"].min()
max_salary=gpt3["Salary"].max()
gpt3["salary_norm"]=gpt3["Salary"].apply(lambda x: (x-min_salary) / (max_salary-min_salary))
min_bonus=gpt3["Bonus"].min()
max_bonus=gpt3["Bonus"].max()
gpt3["bonus_norm"]=gpt3["Bonus"].apply(lambda x: (x-min_bonus) / (max_bonus-min_bonus))
gpt3["Toplam"]=gpt3["salary_norm"]+gpt3["bonus_norm"]
gpt3.sort_values("Toplam",ascending=False).head(5)
```
